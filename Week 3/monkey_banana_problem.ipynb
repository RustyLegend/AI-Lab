```python
from collections import deque
```

```python
print("---------INITIAL STATES---------")
m_position = input("Enter monkey position (A,B,C): ")
b_position = input("Enter Box position (A,B,C): ")
banana_position = input("Enter Banana position (A,B,C): ")
status = 0
has_banana = False
```

```python
q = deque([(m_position, b_position, status, has_banana)])
visited = set([(m_position, b_position, status, has_banana)])
parent = {(m_position, b_position, status, has_banana) : None}
ans_state = []
```

```python
while q:
    state = q.popleft()
    
    if state[3]:
        ans_state = state
        break

    gen_states = []
    
    #move to box 
    if state[2] == 0:
        gen_states.append((state[1], state[1], state[2], state[3]))

    #push box 
    if state[0] == state[1] and state[2] == 0:
        gen_states.append((banana_position, banana_position, state[2], state[3]))

    #climb 
    if state[0] == state[1] and state[2] == 0:
        gen_states.append((state[0], state[1], 1, state[3]))

    #grab
    if state[2] == 1 and state[0] == banana_position: 
        gen_states.append((state[0], state[1], state[2], True))

    for i in gen_states:
        if i not in visited: 
            visited.add(i)
            q.append(i)
            parent[i] = state

print("Monkey at, Box at, Monkey on, Has banana")

if ans_state:
    path = [ans_state]
    i = ans_state
    while parent[i]:
        path.append(parent[i])
        i = parent[i]
    path.reverse()

    for i in path:
        print(i)

    print("Banana Reached")
else:
    print("No solution")
```
