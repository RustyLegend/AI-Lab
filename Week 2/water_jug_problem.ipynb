```python
from collections import deque
```

```python
A = int(input("Enter amount of litres in jug A: "))
B = int(input("Enter amount of litres in jug B: "))
goal = int(input("Enter the the target for jug A: "))
path_exists = False
```

```python
visited = set([(0,0)])
queue = deque([(0,0)])
parent = {(0,0) : None}
ans_state = []

while queue:
    state = queue.popleft()
    
    if state[0] == goal:
        path_exists = True
        ans_state = state
        break
        
    fill_A = (A, state[1])
    fill_B = (state[0], B)
    empty_A = (0, state[1])
    empty_B = (state[0], 0)
    
    new_B = min(state[0], B - state[1])
    A_to_B = (state[0] - new_B, state[1] + new_B)

    new_A = min(state[1], A - state[0])
    B_to_A = (state[0] + new_A, state[1] - new_A)
    
    generated_states = [fill_A, fill_B, empty_A, empty_B, A_to_B, B_to_A]

    for i in generated_states:
        if i not in visited:
            queue.append(i)
            visited.add(i)
            parent[i] = state


if ans_state:
    path = [ans_state]
    i = ans_state
    while parent[i]:
        path.append(parent[i])
        i = parent[i]
    path.reverse()

    for i in path:
        print(i)

    print("Goal Reached")
else:
    print("No solution")
```
